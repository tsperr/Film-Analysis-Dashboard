```python
import pandas as pd 
from sqlalchemy import create_engine
import os
import pymongo
```

```python
movies_df= pd.read_csv("IMDb movies.csv")
movies_df.head(5)
```

```python
clean_movies_df= movies_df[["imdb_title_id", "title", "year", "genre", "country", "budget", "worlwide_gross_income", "production_company"]]
clean_movies_df
```

```python
almost_clean_df = clean_movies_df.dropna(how='any')
almost_clean_df.head(5)
```

```python
final_movie_df = almost_clean_df.loc[(almost_clean_df['year'] == 2000) | (almost_clean_df['year'] == 2001) | 
                                     (almost_clean_df['year'] == 2002) | (almost_clean_df['year'] == 2003) |
                                     (almost_clean_df['year'] == 2004) | (almost_clean_df['year'] == 2005) |
                                     (almost_clean_df['year'] == 2006) | (almost_clean_df['year'] == 2007) |
                                     (almost_clean_df['year'] == 2008) | (almost_clean_df['year'] == 2009) |
                                     (almost_clean_df['year'] == 2010) | (almost_clean_df['year'] == 2011) |
                                     (almost_clean_df['year'] == 2012) | (almost_clean_df['year'] == 2013) |
                                     (almost_clean_df['year'] == 2014) | (almost_clean_df['year'] == 2015) |
                                     (almost_clean_df['year'] == 2016) | (almost_clean_df['year'] == 2017) |
                                     (almost_clean_df['year'] == 2018) | (almost_clean_df['year'] == 2019) |
                                     (almost_clean_df['year'] == 2020) | (almost_clean_df['year'] == 2021),:]

final_movie_df
```

```python
final_movie_df.to_csv("movie_data.csv", encoding='utf-8')
```

```python
# Create connection variable
conn = 'mongodb://localhost:27017'

# Pass connection to the pymongo instance.
client = pymongo.MongoClient(conn)

# Connect to a database. Will create one if not already available.
db = client['project_2']

# Create collection
db_c = db['IMDBdata']

def csv_to_json(filename, header=None):
    data = pd.read_csv(filename, header=header)
    return data.to_dict('records')

db_c.insert_many(csv_to_json('movie_data.csv', header = 0))
```
