```python
clean_data = pd.read_csv("movie_data.csv")
clean_data
```

```python
count_country = clean_data["country"].value_counts()
count_country
```

```python
test = pd.DataFrame(count_country)
test
```

```python
counts = test.reset_index()
counts
```

```python
counts.rename(columns={"index": "name", "country": "count"}, inplace="True")
counts
```

```python
data
```

```python
counts.replace("USA", "United States",inplace=True)
counts.replace("UK","United Kingdom", inplace =True)
counts
```

```python
new_data = pd.merge(counts,data, how='left', on='name')
new_data
```

```python
is_NaN = new_data.isnull()
row_has_NaN = is_NaN.any(axis=1)
rows_with_NaN = new_data[row_has_NaN]

print(rows_with_NaN)
```

```python
map_data = new_data.dropna(how='any')
map_data
```

```python
is_NaN = map_data.isnull()
row_has_NaN = is_NaN.any(axis=1)
rows_with_NaN = map_data[row_has_NaN]

print(rows_with_NaN)
```

```python
map_data.to_csv("static/data/map_data.csv", index = False)
```
